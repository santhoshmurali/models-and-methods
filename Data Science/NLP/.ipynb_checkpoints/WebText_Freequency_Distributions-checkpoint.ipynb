```python
import numpy as np
import pandas as pd
import nltk
from nltk.corpus import webtext
```

```python
webtext.fileids()
```

```python
bow_pirates = webtext.words(fileids='pirates.txt')
```

```python
fdist_pirates = nltk.FreqDist(bow_pirates)
```

```python
distribution_in_df = pd.DataFrame.from_dict(dict(fdist_pirates),orient='index').reset_index()
```

```python
distribution_in_df.columns = ['Key','Val']
```

```python
{fdist_pirates[fdist_pirates.max()]:fdist_pirates.max()}
```
